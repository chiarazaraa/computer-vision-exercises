Answer:
    
i)

-  $l_1 = (1,0,1)$ so $x=-1$ ,   $l_2 = (6,0,2)$ so  $x=-1/3$
- Both lines are vertical and so they are parallel to each other

ii)
- the (3x1) homogenous point vector x should be orthogonal to both $I_1$ and $I_2$ and is so calculated using the cross product:
  $$
l_1 \times l_2 =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
1 & 0 & 1 \\
6 & 0 & 2
\end{vmatrix}
$$

$$
x = (0\cdot2 - 1\cdot0, \; 1\cdot6 - 1\cdot2, \; 1\cdot0 - 0\cdot6) = (0, 4, 0)
$$ 

- The intersection point cannot be plotted as a finite point in a 2D graph:

The image below illustrates the coordinate system employed in this exercise, which is also frequently utilized in computer vision.

Hints:

- The image contains only two color values: 0 and 255, indicating it's a 1-channel black and white image.
- The intersecting lines are straight, parallel to the axis, and have a thickness of one pixel.

```python
# Load image
img = cv2.imread("images/white_cross.png", cv2.IMREAD_GRAYSCALE)
#white pixels
ys, xs = np.where(img == 255)

# vertical line
x_line = int(np.bincount(xs).argmax())

#horizontal line
y_line = int(np.bincount(ys).argmax())

print(f"Intersection point: ({x_line}, {y_line})")

plt.imshow(img, cmap="gray")
plt.scatter(x_line, y_line, color="pink", marker="*", s=100, label="Intersection")
plt.legend()
plt.title("White Cross Intersection")
plt.show()
```

## Ex 1.3 Singular Value Decomposition (SVD) (3 points)

Singular Value Decomposition (SVD): Every $m \times p$ matrix factors into $A_{n \times p} = U_{n \times n} \Sigma_{n \times p} V^T_{p \times p}$

Singular Value Decomposition can be used for image compression by first applying SVD to the image matrix and then recomposing it by using only some of the singular values contained in matrix $\Sigma_{n \times p}$ and dropping the respective columns and rows in matrices $U_{n \times n}$ and $V^T_{p \times p}$ in order to make the matrices compatible for recomposing the image matrix.

Try this for yourself by decomposing images/lenna.png (open the image as grayscale) with SVD, and then recompose the image by using only the first 50 singular values (you can use the ready-made SVD function in numpy.linalg library, no need to implement it from scratch unless you want to). Also try using different number of singular values. How many singular values are needed before the image no longer appears blurry (Optional extra: how is the image size changed, i.e how much memory you save by compressing the image)?

Explain why this method of image compression works, what is the role of singular values and what information they contain? Brief explanation is enough for this assignment.

### Answer

- This method works because most of the important information in an image is stored in the largest singular values.
- The singular values in the matrix $\Sigma$ show how much importance each part of the image has. Large singular values capture the main structures, edges, and shapes of the image, while small singular values contain fine details or noise. By keeping only the largest singular values and removing the small ones, we can compress the image and still keep it looking almost the same. (As in this example, starting from about $k=80$ the image already looks clear and no longer blurry, which means that the first 80 singular values are enough to represent the important information of the image.)

```python
img = cv2.imread("images/lenna.png", cv2.IMREAD_GRAYSCALE)

#SVD
U, S, Vt = np.linalg.svd(img, full_matrices=False)

#reconstruct the image using k singular values
def reconstruct_mg(U, S, Vt, k):
    return np.dot(U[:, :k], np.dot(np.diag(S[:k]), Vt[:k, :]))

# k values (
k_values = [10, 30, 50, 80, 100, 200]
min_dim = min(img.shape)
k_values = [min(k, min_dim) for k in k_values] 

# Plot 
plt.figure(figsize=(15,5))
for i, k in enumerate(k_values):
    plt.subplot(1, len(k_values), i+1)
    plt.imshow(reconstruct_mg(U, S, Vt, k), cmap='gray')
    plt.title(f'k={k}')
    plt.axis('off')
plt.show()

# Memory calculation for k=50
k = min(50, min_dim)  # ensure k does not exceed image dimensions
compressed_size = k * (U.shape[0] + Vt.shape[1] + 1)
original_size = img.size
print(f"Original size: {original_size} pixels")
print(f"Compressed size with k={k}: {compressed_size} pixels")
print(f"Memory saved: {100*(1 - compressed_size/original_size):.2f}%")
```
